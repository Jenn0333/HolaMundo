```python
#Apartado 1.- INTERFAZ
from abc import ABC
from abc import abstractmethod
class Gato(ABC):
    @abstractmethod
    def meow(self):
        pass

#Apartado 2.- RAZAS GATO
class GatoNaranja(Gato):
  def meow(self):
        print("Donde está el puto churu </3");

class GatoPersa(Gato):
  def meow(self):
        print("Meow solo");


gNaranja = GatoNaranja();
gPersa = GatoPersa();

#Apartado 3.- MAIN, GATO
print("¿Qué tipo de gato quieres?");
tipo=input();
print("Has elegido el tipo de gato: "+tipo)

if(tipo=="Naranja"):
  gNaranja.meow();
elif(tipo=="Persa"):
  gPersa.meow();
else:
    print("no")
```
